```python
df2 = df.groupby('Continent').mean(numeric_only=True)

df2
```

```python
df2 = df2.T

df2
```

```python
rows_to_drop = ['Rank', 'Area (km²)', 'Density (per km²)', 'Growth Rate', 'World Population Percentage']
df_cleaned = df2.drop(index=rows_to_drop)
df_cleaned
```

```python
df_reversed = df_cleaned.iloc[::-1]

df_reversed
```

```python
df_reversed.plot(title='Population Growth by Continent', marker='o', figsize=(20,10))
plt.xlabel('Year')
plt.ylabel('Population')
plt.legend(title='Continent')
plt.grid()
plt.show()
```
